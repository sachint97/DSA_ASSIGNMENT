Question 1
Given an integer array nums of length n and an integer target, find three integers
in nums such that the sum is closest to the target.
Return the sum of the three integers.

You may assume that each input would have exactly one solution.

Example 1:
Input: nums = [-1,2,1,-4], target = 1
Output: 2

Explanation: The sum that is closest to the target is 2. (-1 + 2 + 1 = 2).

```python
def threeSumClosest(nums, target):
        nums.sort()
        i=0
        l=len(nums)
        diff = float('inf')
        sol=0
        while i < l:
            start = i+1
            end = l-1
            while start<end:
                summ = nums[i]+nums[start]+nums[end]
                if abs(target - summ)<diff:
                    diff = abs(target - summ)
                    sol=summ
                elif summ < target:
                    start += 1
                else:
                    end -= 1
            i+=1
        return sol

nums = [-1,2,1,-4]
target = 1
threeSumClosest(nums,target)
```

Question 2
Given an array nums of n integers, return an array of all the unique quadruplets
[nums[a], nums[b], nums[c], nums[d]] such that:
           ‚óè 0 <= a, b, c, d < n
           ‚óè a, b, c, and d are distinct.
           ‚óè nums[a] + nums[b] + nums[c] + nums[d] == target

You may return the answer in any order.

Example 1:
Input: nums = [1,0,-1,0,-2,2], target = 0
Output: [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]

```python
def fourSum(nums, target):
  nums.sort()
  i=0
  l=len(nums)
  sol=set()
  while i < l-3:
    j = i+1
    while j < l-2:
      start = j+1
      end = l-1
      while start<end:
        summ = nums[i]+nums[j]+nums[start]+nums[end]
        if summ == target:
            sol.add((nums[i],nums[j],nums[start],nums[end]))
            start += 1
        elif summ < target:
            start += 1
        else:
              end -= 1
      j+=1
    i+=1
  return list(list(x) for x in sol)

nums = [1,0,-1,0,-2,2]
target = 0 
fourSum(nums,target)
```

The next permutation of an array of integers is the next lexicographically greater

If such an arrangement is not possible, the array must be rearranged as the

Given an array of integers nums, find the next permutation of nums.
The replacement must be in place and use only constant extra memory.

**Example 1:**
Input: nums = [1,2,3]
Output: [1,3,2]

```python
def nextPermutation(nums):
  l=len(nums)
  i=l-1
  peak=0
  while i>0 :
      if nums[i]>nums[i-1]:
          peak = i
          break
      i-=1
  if peak == 0: # check if list is in descending order
      return nums.sort()
  j=peak+1
  while j<l:
      if nums[j]<nums[i] and nums[j]>nums[i-1]:
          peak = j
      j+=1
  temp = nums[i-1]
  nums[i-1]=nums[peak]
  nums[peak] = temp
  sor=sorted(nums[i:])
  nums[i:]=sor
  return nums

nums = [1,2,3]
nextPermutation(nums)
```

üí° **Question 5**
You are given a large integer represented as an integer array digits, where each
digits[i] is the ith digit of the integer. The digits are ordered from most significant
to least significant in left-to-right order. The large integer does not contain any
leading 0's.

Increment the large integer by one and return the resulting array of digits.

**Example 1:**
Input: digits = [1,2,3]
Output: [1,2,4]

**Explanation:** The array represents the integer 123.
Incrementing by one gives 123 + 1 = 124.
Thus, the result should be [1,2,4].

</aside>

```python
def plusOne(digits):
  l=len(digits)
  i = l-1
  carry = 0
  while i>=0:
    digits[i]+=carry
    carry=0
    if i==l-1:
        digits[i]+=1
    if digits[i]>9:
        carry = 1
        digits[i]=0
    i-=1
  if carry:
    digits.insert(0,carry)
  return digits

digits = [1,2,3]
plusOne(digits)
```

Question 6
Given a non-empty array of integers nums, every element appears twice except
for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only
constant extra space.

Example 1:
Input: nums = [2,2,1]
Output: 1

```python
def singleNumber(nums):
  res = nums[0]
  for val in nums[1:]:
      res = res ^ val
  return res
nums = [2,2,1]
singleNumber(nums)
```

Question 7
You are given an inclusive range [lower, upper] and a sorted unique integer array
nums, where all elements are within the inclusive range.

A number x is considered missing if x is in the range [lower, upper] and x is not in
nums.

Return the shortest sorted list of ranges that exactly covers all the missing
numbers. That is, no element of nums is included in any of the ranges, and each
missing number is covered by one of the ranges.

Example 1:
Input: nums = [0,1,3,50,75], lower = 0, upper = 99
Output: [[2,2],[4,49],[51,74],[76,99]]

Explanation: The ranges are:
[2,2]
[4,49]
[51,74]
[76,99]

```python
def missing_range(nums,lower,upper):
  i=0
  l=len(nums)
  m_ranges=[]
  while i<l:
    if i==0 and nums[i]!=lower:
      m_ranges.append([lower,nums[i]-1])
    elif i==l-1 and nums[i]!=upper:
      m_ranges.append([nums[i]+1,upper])
    elif nums[i+1]-nums[i]!=1:
      m_ranges.append([nums[i]+1,nums[i+1]-1])
    i+=1
  return m_ranges

nums = [0,1,3,50,75]
lower = 0
upper = 99 
missing_range(nums,lower,upper)

```

Question 8
Given an array of meeting time intervals where intervals[i] = [starti, endi],
determine if a person could attend all meetings.

Example 1:
Input: intervals = [[0,30],[5,10],[15,20]]
Output: false

```python
def meeting_time(nums):
  l=len(nums)
  i=0
  while i<l-1:
    j=i+1
    while j<l:
      if nums[j][0]>nums[i][0] and nums[j][0]<nums[i][1]:
        return False
      j+=1
    i+=1
  return True

intervals = [[0,30],[5,10],[15,20]]
meeting_time(intervals)
```
